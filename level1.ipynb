__[문제5]__

str = "야호야호야호호야호야야오햐오햐오햐오"

str을 3가지 이상의 다른 문자열 슬라이싱을 이용하여 확인해보세요!

출력 예)

처음 5글자: 야호야호야


```python
str_var = "야호야호야호호야호야야오햐오햐오햐오"


print("처음 5글자:", str_var[:5])
print("마지막 4글자:", str_var[-4:])
print("2번째부터 6번째까지:", str_var[1:6])

```
